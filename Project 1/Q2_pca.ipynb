```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import decomposition, datasets
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("C:/Users/DELL/Desktop/DM/dist1_500_1.txt",header= None,sep = " ")
```

```python
df1 = pd.read_csv("C:/Users/DELL/Desktop/DM/dist1_500_2.txt",header= None,sep = " ")
```

```python
df2 = pd.read_csv("C:/Users/DELL/Desktop/DM/dist2_500_1.txt",header= None,sep = " ")
```

```python
df3 = pd.read_csv("C:/Users/DELL/Desktop/DM/dist2_500_2.txt",header= None,sep = " ")
```

```python

dataset1=df.append(df1)
dataset2=df1.append(df2)
clean_1=dataset1.dropna(axis='index')
clean_2=dataset2.dropna(axis='index')
X=clean_1
X.shape
Y=clean_2
```

```python
sc = StandardScaler()
```

```python
X_std = sc.fit_transform(X)
Y_std= sc.fit_transform(Y)
```

```python
pca = decomposition.PCA(n_components=2)
```

```python
X_std_pca = pca.fit_transform(X_std)
Y_std_pca = pca.fit_transform(Y_std)
```

```python
X_std_pca.shape
```

```python
Y_std_pca.shape
```

```python
Y_std_pca
```

```python
explained_variance = pca.explained_variance_
```

```python
explained_variance
Total_variance=0
for index,value in enumerate(explained_variance):
    Total_variance+=value

print(Total_variance,"%")
```

```python
print('Original data plotted:')
plt.plot(X)
plt.show()
```

```python
print('Transformed Data:')
plt.plot(X_std_pca)
plt.show()
```
